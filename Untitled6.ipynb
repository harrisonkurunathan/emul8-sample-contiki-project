```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate vehicle movement
def simulate_vehicle_movement(num_vehicles, num_steps):
    movements = []
    for _ in range(num_vehicles):
        x, y = np.random.rand(2) * 100  # Random initial position
        trajectory = []
        for _ in range(num_steps):
            trajectory.append([x, y])
            x += (np.random.rand() - 0.5) * 2  # Random movement in x
            y += (np.random.rand() - 0.5) * 2  # Random movement in y
        movements.append(trajectory)
    return np.array(movements)

num_vehicles = 10
num_steps = 100
data = simulate_vehicle_movement(num_vehicles, num_steps)

# Visualize the movements
for vehicle in data:
    plt.plot(vehicle[:, 0], vehicle[:, 1], marker='o')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Vehicle Movements')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

def prepare_data(data, time_steps):
    X, y = [], []
    for vehicle in data:
        for i in range(len(vehicle) - time_steps):
            X.append(vehicle[i:i+time_steps])
            y.append(vehicle[i+time_steps])
    return np.array(X), np.array(y)

time_steps = 5
X, y = prepare_data(data, time_steps)

# Normalize the data
scaler = MinMaxScaler()
X = scaler.fit_transform(X.reshape(-1, X.shape[-1])).reshape(X.shape)
y = scaler.transform(y)

print("Input shape:", X.shape)
print("Output shape:", y.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Build the LSTM model
model = Sequential()
model.add(LSTM(50, input_shape=(time_steps, 2)))
model.add(Dense(2))
model.compile(optimizer='adam', loss='mse')

# Train the model
history = model.fit(X, y, epochs=50, batch_size=64, validation_split=0.2)

# Plot training history
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Predict future positions
def predict_future_positions(model, initial_positions, time_steps, future_steps):
    future_positions = initial_positions.copy()
    for _ in range(future_steps):
        input_data = future_positions[-time_steps:]
        input_data = scaler.transform(input_data)
        input_data = input_data[np.newaxis, ...]
        next_position = model.predict(input_data)
        next_position = scaler.inverse_transform(next_position)
        future_positions = np.vstack([future_positions, next_position])
    return future_positions

# Example of prediction
vehicle_index = 0
initial_positions = data[vehicle_index][:time_steps]
predicted_positions = predict_future_positions(model, initial_positions, time_steps, num_steps-time_steps)

# Plot actual vs predicted positions
plt.plot(data[vehicle_index][:, 0], data[vehicle_index][:, 1], marker='o', label='Actual')
plt.plot(predicted_positions[:, 0], predicted_positions[:, 1], marker='x', label='Predicted')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Actual vs Predicted Vehicle Movements')
plt.legend()
plt.show()
```
